```python
input_path = "data/paris.jpg"
input_image = plt.imread(input_path)
univ, output = segment(input_image, sigma, 2000, 100)
```

```python
h,w = input_image.shape[:2]
```

```python
bb = {str(comp) : np.array([[w-1,0],[w*h-1,0]]) for comp in np.unique(univ.elts[:,2])}
```

```python
for y in range(h):
    for x in range(w):
        pixel_id = y*w+x
        comp = str(univ.find(pixel_id))
        if bb[comp][0][0]%w > pixel_id%w: # most left
            bb[comp][0][0] = pixel_id
        if bb[comp][0][1]%w < pixel_id%w: # most right
            bb[comp][0][1] = pixel_id
        if bb[comp][1][0] > pixel_id: # most up
            bb[comp][1][0] = pixel_id
        if bb[comp][1][1] < pixel_id: # most down
            bb[comp][1][1] = pixel_id
```
